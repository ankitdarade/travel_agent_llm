```python
import requests

# Seting the new API key for Geoapify
geoapify_api_key = "6f047acfd32c469aa65f0e957e329b2e" 

def get_coordinates(city: str):
    """Fetches the geographical coordinates (latitude and longitude) for a given city."""
    url=f"https://api.geoapify.com/v1/geocode/search?text={city}&apiKey={geoapify_api_key}"
    response=requests.get(url)
    
    print("Status Code for Coordinates:", response.status_code)  # Debugging line
    try:
        data=response.json()
        print("Raw JSON Response for Coordinates:", data)  # Debugging line

        lat=data["features"][0]["geometry"]["coordinates"][1]
        lon=data["features"][0]["geometry"]["coordinates"][0]
        return lat, lon
    except Exception as e:
        print("Error parsing coordinates:", e)
        return None, None

def get_attractions(city: str):
    """Fetches the top tourist attractions near the given city using Geoapify API."""
    lat, lon=get_coordinates(city)
    if lat is None or lon is None:
        return ["Unable to fetch coordinates for the city."]
    
    url=f"https://api.geoapify.com/v2/places?categories=tourism.attraction&filter=circle:{lon},{lat},10000&limit=10&apiKey={geoapify_api_key}"
    
    # Fetch data from Geoapify API
    response=requests.get(url)
    
    print("Status Code for Attractions:", response.status_code)  # Debugging line
    try:
        data = response.json()
        print("Raw JSON Response for Attractions:", data)  # Debugging line

        # Check if 'features' exist and if they contain any places
        places=data.get("features", [])
        if not places:
            print("No attractions found in this area.")
            return ["No attractions found."]
        
        # Extracting place names
        attractions=[place["properties"].get("name", "Unnamed Place") for place in places]
        return attractions
    
    except Exception as e:
        print("Error parsing attractions:", e)
        return ["Error fetching or parsing data."]

# Example Usage:
city="Jaipur"
attractions=get_attractions(city)

print("Top Attractions:", attractions)
```

```python
import requests
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# our all apikeys 
HUGGINGFACE_MODEL="google/flan-t5-base"
WEATHER_API_KEY="e9a58064035e40855a57adc8dcb7d1e0"
GEOAPIFY_API_KEY="6f047acfd32c469aa65f0e957e329b2e"

#starting model
print("[INFO] Loading FLAN-T5 model...")
tokenizer=AutoTokenizer.from_pretrained(HUGGINGFACE_MODEL)
model=AutoModelForSeq2SeqLM.from_pretrained(HUGGINGFACE_MODEL)

# taking data

def get_weather(city: str):
    url="http://api.openweathermap.org/data/2.5/weather"
    params={"q": city, "appid": WEATHER_API_KEY, "units": "metric"}
    response=requests.get(url, params=params)
    if response.status_code == 200:
        data=response.json()
        desc=data["weather"][0]["description"]
        temp=data["main"]["temp"]
        return f"{desc}, {temp}°C"
    return "Weather data not available."

def get_attractions(lat: float, lon: float, radius: int = 10000):
    url="https://api.geoapify.com/v2/places"
    params={
        "categories":"tourism.attraction",
        "filter":f"circle:{lon},{lat},{radius}",
        "limit":10,
        "apiKey":GEOAPIFY_API_KEY
    }
    response=requests.get(url, params=params)
    if response.status_code==200:
        data=response.json()
        return [place["properties"].get("name", "Unnamed") for place in data.get("features", [])]
    return ["Attractions data not available."]

# iternatory genarator
def generate_itinerary(city:str,lat:float,lon:float,start_date:str,days:int=3):
    print("[INFO] Gathering weather and attractions...")
    weather=get_weather(city)
    attractions=get_attractions(lat, lon)
    attraction_list=', '.join(attractions)

    prompt=f"""
Create a detailed {days}-day travel itinerary for a visitor to {city}, starting on {start_date}.
Weather forecast:{weather}.
Top attractions:{attraction_list}.

The itinerary should have:
- A clear structure: Day 1, Day 2, Day 3
- For each day, list:
  - Morning: [activity]
  - Afternoon: [activity]
  - Evening: [activity]
Make it vivid, engaging, and customized to Jaipur's culture and highlights.
"""

    print("[INFO] Generating itinerary with LLM...")
    inputs=tokenizer(prompt, return_tensors="pt", truncation=True)
    outputs=model.generate(
        **inputs,
        max_length=800,
        do_sample=True,
        temperature=0.7,
        top_p=0.92,
        top_k=50,
        num_beams=3,
        no_repeat_ngram_size=2
    )
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return decoded


# runiung by taking sample 
if __name__ == "__main__":
    city="Jaipur"
    lat=26.9124
    lon=75.7873
    start_date="2025-05-01"

    itinerary=generate_itinerary(city, lat, lon, start_date)
    print("\nGenerated Itinerary:\n")
    print(itinerary)
```
